```python
import cv2

def applyclahe(imgpath,imgname):
    bgr = cv2.imread(imgpath)
    lab = cv2.cvtColor(bgr, cv2.COLOR_BGR2LAB)
    lab_planes = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
    lab_planes[0] = clahe.apply(lab_planes[0])
    lab = cv2.merge(lab_planes)
    bgr = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
    cv2.imwrite(imgname,bgr)

for k in range(21,41):
    imgpath='D:/google drive/Missions/mission Retina/navdeep sir datasets/DRIVE/training/maskedimgs/{}.jpg'.format(k)
    finimg='D:/google drive/Missions/mission Retina/navdeep sir datasets/DRIVE/training/afterclaheimgs/'+str(k)+'.jpg'
    applyclahe(imgpath,finimg)
```
